# Retina Analysis

```python
import scanpy as sc
import scvi
import anndata
from sklearn.cluster import KMeans
```

```python
adata=scvi.data.retina()
```

```python
adata.obs
```

```python
sc.pp.calculate_qc_metrics(adata,qc_vars=[],percent_top=[],log1p=False,inplace=True)
```

```python
sc.pl.scatter(adata,"n_genes_by_counts","total_counts")
```

### Normalization

```python
adata.layers["counts"]=adata.X.copy()
sc.pp.normalize_total(adata,target_sum=1000)
sc.pp.log1p(adata)
adata.raw=adata
```

```python
sc.pp.combat(adata, key='batch', covariates=None, inplace=True)
```

```python
sc.pp.highly_variable_genes(
    adata,
    n_top_genes=2000,
    # subset=True, # to automatically subset to the 4000 genes
    layer="counts",
    flavor="seurat_v3"
)
```

```python
adata.var
adata=adata[:,adata.var.highly_variable].copy()
```

```python
sc.pp.scale(adata)
sc.pp.pca(adata)
```

```python
sc.pp.neighbors(adata,n_neighbors=20)
```

```python
sc.tl.umap(adata)
```

```python
sc.pl.umap(adata, color=["labels","batch"])
```

```python
adata.obs.nunique()
```

```python
X_pca = adata.obsm['X_pca'] 
kmeans = KMeans(n_clusters=8, random_state=0).fit(X_pca) 
adata.obs['kmeans'] = kmeans.labels_.astype(str)
sc.pl.umap(adata,color=["labels",'kmeans'])
```
